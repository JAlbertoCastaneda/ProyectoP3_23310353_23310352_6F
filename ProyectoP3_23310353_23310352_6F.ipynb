```python
from ultralytics import YOLO

# 1. Cargamos un modelo base ligero (YOLOv8 Nano)
model = YOLO('yolov8n.pt')

# 2. Empezamos el entrenamiento apuntando al archivo data.yaml del dataset
results = model.train(
    data=f"{dataset.location}/data.yaml",
    epochs=50,       # 50 vueltas de entrenamiento para empezar
    imgsz=640,       # Tamaño de las imágenes
    plots=True       # Para que nos genere gráficas de rendimiento
)
```

```python
model_entrenado = YOLO('runs/detect/train/weights/best.pt')


resultados = model_entrenado.predict(source=f"{dataset.location}/valid/images", save=True, imgsz=640)
```
